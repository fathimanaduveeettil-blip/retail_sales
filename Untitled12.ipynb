```python
month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

monthly_sales = (
    sales.groupby('Month_Name')['Sales']
    .sum()
    .reindex(month_order)
)

print(monthly_sales)
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
yearly_sales = sales.groupby('Year')['Sales'].sum()

print(yearly_sales)
```

```python
yoy_growth = yearly_sales.pct_change() * 100

print(yoy_growth)
```

```python
plt.figure(figsize=(8,5))

yoy_growth.plot(kind='bar')

plt.title('Year-over-Year Growth (%)')
plt.xlabel('Year')
plt.ylabel('Growth (%)')
plt.grid(axis='y')

plt.show()
```

```python
yoy_table = yearly_sales.to_frame(name='Sales')

yoy_table['YoY Growth %'] = yoy_growth

print(yoy_table)
```

```python
import matplotlib.pyplot as plt

yoy_growth.plot(marker='o', figsize=(8,5))

plt.title('Year-over-Year Sales Growth')
plt.xlabel('Year')
plt.ylabel('Growth (%)')
plt.grid(True)

plt.show()
```

```python
total_revenue = sales['Sales'].sum()

print("Total Revenue =", total_revenue)
```

```python
total_profit = sales['Profit'].sum()

print("Total Profit =", total_profit)
```

```python
profit_margin = (total_profit / total_revenue) * 100

print("Profit Margin =", round(profit_margin,2), "%")
```

```python
total_orders = sales['OrderID'].nunique()

print("Total Orders =", total_orders)
```

```python
aov = total_revenue / total_orders

print("Average Order Value =", round(aov,2))
```

```python
top_customers = (
    sales.groupby('CustomerName')['Sales']
         .sum()
         .sort_values(ascending=False)
         .head(10)
)

print(top_customers)
```

```python
plt.figure(figsize=(10,5))

top_customers.plot(kind='bar')

plt.title('Top 10 Customers')
plt.xlabel('Customer')
plt.ylabel('Sales')

plt.show()
```

```python
returned_orders = sales[sales['Returned'] == 'Yes'].shape[0]

total_orders = sales.shape[0]

return_rate = (returned_orders / total_orders) * 100

print("Return Rate =", round(return_rate,2), "%")
```

```python
customer_orders = sales.groupby('CustomerID')['OrderID'].nunique()

repeat_customers = (customer_orders > 1).sum()

total_customers = customer_orders.count()

repeat_rate = (repeat_customers / total_customers) * 100

print("Repeat Customer Rate =", round(repeat_rate,2), "%")
```

```python
print("Total Revenue:", total_revenue)
print("Total Profit:", total_profit)
print("Profit Margin:", round(profit_margin,2), "%")
print("Average Order Value:", round(aov,2))
print("Return Rate:", round(return_rate,2), "%")
print("Repeat Customer Rate:", round(repeat_rate,2), "%")
```
